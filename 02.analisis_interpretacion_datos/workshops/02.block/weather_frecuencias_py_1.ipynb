```python
###############################################################
# TABLAS DE FRECUENCIAS Y HISTOGRAMAS EN PYTHON
# Usando el dataset weather.csv
###############################################################

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# 1. Cargar datos
df = pd.read_csv("weather.csv")
# 1. Primeras filas
print(df.head())

# 2. Informacin general: tipos de variable, nulos
print(df.info())

# 3. Resumen estadstico para numricas
print(df.describe())

# 4. Nulos por columna
print(df.isnull().sum())

# 5. Total de nulos
print(df.isnull().sum().sum())

# 6. Tipos de datos de cada columna
print(df.dtypes)

# 2. Seleccionar variable numrica
x = df["MaxTemp"].dropna()

# -------------------------------------------------------------------
# TABLA DE FRECUENCIAS NO AGRUPADAS
# -------------------------------------------------------------------

# Frecuencia absoluta
freq_abs = x.value_counts().sort_index()

# Frecuencia relativa
freq_rel = freq_abs / len(x)

# Frecuencia acumulada
freq_acum = freq_abs.cumsum()

# Tabla final
tabla = pd.DataFrame({
    "Valor": freq_abs.index,
    "Frecuencia": freq_abs.values,
    "Relativa": freq_rel.round(3).values,
    "Acumulada": freq_acum.values
})

print("\nTABLA DE FRECUENCIAS")
print(tabla)

```
