2. **¿Qué es la estadística?**

- Desde un punto de vista mas amplio, definir la estadística como la ciencia que estudia como debe emplearse la informacion y como dar una guía de accion en situaciones practicas que entrañan incertidumbre.
- La **Estadística** se ocupa de los **metodos y procedimientos** para **recoger, clasificar, resumir, hallar regularidades y analizar los datos**, siempre y cuando la variabilidad e incertidumbre sea una causa intrínseca de los mismos; así como de realizar inferencias a partir de ellos, con la finalidad de ayudar a la toma de decisiones y en su caso formular predicciones.
- La estadística es la ciencia de aprender de los datos, desde su recolección hasta la interpretación.
>  - Imaginen que queremos saber qué comida prefieren los estudiantes en la cafetería. Si solo registramos los datos, eso es **"recopilación"**, pero cuando **los analizamos** para identificar la comida favorita, eso es **estadística**.

  - Podríamos por tanto **clasificar la Estadística** en:
    - **Estadística descriptiva:** *Cuando los resultados del análisis no pretenden ir más allá del conjunto de datos*
      - Describe, analiza y representa un grupo de datos utilizando métodos numéricos y gráficos que resumen y presentan la información contenida en ellos.
      - Con la que  nos limitamos a describir una población  basandonos en la informacion recogida de muestra.
    - **Estadística inferencial:** *Cuando el objetivo del estudio es derivar las conclusiones obtenidas a un conjunto de datos más amplio*.
      - Apoyándose en el cálculo de probabilidades y a partir de datos muestrales, efectúa estimaciones, decisiones, predicciones u otras generalizaciones sobre un conjunto mayor de datos.
      - Con la que pretendemos establecer conclusiones sobre la población.

3. **Población, muestra y muestreo**

La definición de **estadística** emplea primeramente el concepto de **dato**, que no solo es un número, sino un número en un **contexto**, con lo cual es **información** recolectada sobre algo. Pero ese *«algo»* es lo que llamaremos **individuo** el cual conforma un colectivo que llamamos **población**, que es finalmente sobre lo que nos interesa estudiar y sacar conclusiones. Por lo tanto, la estadística no se encarga de cualquier fenómeno, sino de aquellos que son colectivos y que no atienden a leyes deterministas (de las cuales se encargan las ciencias exactas), es decir, de aquellos que contienen algún elemento de **incertidumbre**.

El proceso mediante el cual seleccionamos a los individuos que van a formar parte de la **muestra** se denomina **muestreo** y es clave para garantizar un mínimo de calidad en los datos obtenidos *(es decir, una información importante sobre la población)*, que ayude a validar futuros análisis y conclusiones. Lo deseable al recoger la muestra es que los individuos seleccionados configuren una **muestra representativa** de su **población**, es decir, que contenga una diversidad muy similar a la de la población de origen.



Siempre que obtengamos una muestra estamos expuestos al **error de muestreo**, producto de inferir o extrapolar a partir de un trozo de realidad *(la muestra)*, el cómo será la realidad entera *(la población)*. La clave será reducir este error, inherente al propio proceso de **muestreo**, al mínimo.



El proceso de extrapolar las características y propiedades de la muestra a las de la población se conoce como **inferencia estadística** y, dada su importancia, ha devenido en una rama de la estadística *(generalmente se habla de estadística descriptiva y de la inferencial)*.

> *- Como el estudio anterior son en realidad todos los estudios que se llevan hoy en día en España, pues **los censos** o registros exhaustivos a toda la población ya no se practican desde el año 2000 cuando tuvo lugar el último censo de la población española.*

4. **Tipos de variables estadísticas**

Tal y como observamos en el esquema existen dos tipos de *variables estadísticas*: las **categóricas** y **cuantitativas**. La primera de ellas está dividida a su vez en dos *clases*, dependiendo de si las *categorías* son meramente cualitativas, son las llamadas **nominales**, o si además poseen orden, las **ordinales**. Las *cuantitativas* pueden ser **discretas** cuando toman un número finito de valores o **continuas** cuando pueden tomar infinitos valores como por ejemplo *las magnitudes físicas (altura, peso, etc.)*



La clasificación anterior de los tipos de variables no es única. Otros autores las subdividen de otro modo, aunque este es probablemente el más común. También podemos **clasificar las variables según su enfoque metodológico**:
- Variables dependientes.
- Variables independientes.

> ***Ejemplo:** aprobado en Lengua en el 1er Cuatrimestre será variable dependiente de otra independiente como puede ser el número de horas de estudio de Lengua. Se supone que pretendemos explicar el hecho de aprobar Lengua a partir del número de horas estudiadas para la asignatura, lo cual parece razonable (aunque existirán otros factores).*

- *La variable dependiente, respuesta y explicada por un lado, e independiente, explicativa y predictora por el otro, no son más que sinónimos de un mismo rol que desempeña la variable. En economía u otras disciplinas pueden emplearse otros términos equivalente como variables endógenas y exógenas, etc.*
- *Otro tipo de variable al que conviene ponerle nombre es el de las **variables intermediarias u omitidas**, variables que no son contempladas por el estudio o el modelo planteado en cuestión, pero que en el fondo estarían actuando de variables explicativas de nuestra variable dependiente, pero de un modo digamos oculto, o mejor dicho «desde la sombra». Conviene identificarlas para no establecer asociaciones y presuponer causalidades infundadas.*
- *Otro tipo de variable muy empleado en estadística es el de las **variables dicotómicas**, ya que son muy útiles para describir el hecho de que ocurra **algo (1)** o **no ocurra (0)**.*
- ***En la práctica una misma variable puede ser recodificada de diferentes modos**, como por ejemplo la variable edad. En teoría se trata de una variable continua (la edad es el tiempo pasado desde el nacimiento, que es una magnitud continua), sin embargo, puede ser recogida en su dimensión puramente categórica ordinal si solo apuntamos o codificamos los intervalos de edad, tal y como ocurre en numerosas encuestas. **(Ejemplo: Menor de edad- De 18 a 25 años- Mayor de 25)**.*

- **Ejemplos de cada tipo de variable son:**
  - **Categórica nominal** es el género, el grupo al que pertenecen los alumnos, etc.
  - **Categórica ordinal** es el curso al que pertenecen los alumnos (Ejemplo: 1ºESO, 2ºESO,…, 2ºBachillerato).
  - **Cuantitativa discreta** es el número de asignaturas suspensas en un cuatrimestre.
  - **Cuantitativa continua** es el tiempo empleado en hacer el examen.

7. **Representando los datos: distribución de frecuencias**

- Ahora vamos a pasar la fase de organización y representación de datos. Lo primero que se nos ocurre hacer con los datos es contarlos. Anotar sus repeticiones, es decir, el número de veces que se repite un valor o una categoría de una variable. A estas magnitudes las llamamos **frecuencias**.

**Clasificamos las frecuencias de la siguiente manera:**
- Las **absolutas**, que denotamos **$n_i$** donde la **$i$** hace referencia a la categoría o valor *i-ésimo* de la variable (también llamado **modalidad**).
- Las **relativas** que se obtienen como las absolutas en relación al **N total** o suma de todas las frecuencias absolutas de todas las modalidades, que en realidad no es más que el tamaño de la muestra:
  - $f_i = \frac{n_i}{N}$, siendo $N = \sum_{i=1}^{k} n_i$

11. **Retos de la estadística en el Big Data**

La **estadística** es una disciplina clásica cuya definición como “ciencia que recolecta y analiza los datos” proviene del siglo XIX.  
Con la llegada de los **computadores**, **Internet** y el **Big Data**, los entornos con los que trabaja la estadística han cambiado radicalmente.

Antes se manejaban conjuntos relativamente pequeños de datos; hoy la cantidad de información disponible supera nuestra capacidad de análisis.  
Esto genera un nuevo problema: **“tenemos tantos datos que no hay manera de analizarlos”**.  

A pesar de tener más información que nunca, a menudo **no logramos extraer conocimiento útil**.  
Por ello, la estadística debe adaptarse y desarrollar nuevos métodos que permitan **aprender de los datos masivos** disponibles.

**Retos Principales de la Estadística en Entornos Big Data**

1. **Excesiva Cantidad de Información y Datos**
- Los métodos estadísticos clásicos no fueron diseñados para manejar grandes volúmenes de información, por lo que pueden resultar **ineficientes** en entornos Big Data.  
  - Esto requiere:
    - Diseñar **códigos y algoritmos eficientes** que permitan aplicar métodos clásicos.  
    - Desarrollar **nuevos enfoques estadísticos** capaces de trabajar con grandes cantidades de datos.  

Además, surge el problema de los **outliers (valores atípicos)**:  
En conjuntos pequeños pueden eliminarse, pero en Big Data, los outliers pueden representar **porciones significativas del conjunto**, por lo que **no deben ignorarse**.

> **Ejemplo:** En redes sociales, comportamientos "atípicos" de usuarios pueden ser claves para detectar fraudes o tendencias emergentes.

**Efecto Hawthorne**

El **Efecto Hawthorne** se refiere al cambio en el comportamiento de las personas cuando son conscientes de que están siendo observadas.  
Este fenómeno fue descubierto entre 1924 y 1932 en los estudios realizados en la planta **Hawthorne Works** de la empresa **Western Electric** (Chicago, EE. UU.).

- **Descripción**
Durante los experimentos, los investigadores modificaron variables como la iluminación, las pausas y la duración de la jornada laboral para analizar su impacto en la productividad.  
Sorprendentemente, **la productividad aumentaba independientemente de las condiciones**, lo que llevó a concluir que la simple atención de los investigadores generaba una mejora temporal en el desempeño.

- **Conclusión**
El Efecto Hawthorne demuestra que:
  - Las personas tienden a mejorar su rendimiento cuando saben que están siendo observadas.  
  - La **motivación** y el **reconocimiento** pueden influir tanto como los factores materiales o ambientales.  
  - En contextos laborales o de investigación, es importante considerar este efecto para evitar **sesgos** en los resultados.

- **Aplicaciones**
  - **Gestión de personal:** la supervisión visible puede aumentar la productividad temporalmente.  
  - **Investigación social y psicológica:** debe tenerse en cuenta para evitar que la observación afecte el comportamiento del grupo.  
  - **Análisis de datos:** puede introducir **ruido** o **sesgos** en experimentos o pruebas A/B.

- **Ejemplo práctico**
En un experimento moderno, si los empleados saben que sus métricas están siendo monitoreadas, podrían trabajar más rápido o con más cuidado, **no porque cambien sus condiciones**, sino por la conciencia de estar siendo evaluados.

**Series Temporales**

Una **serie temporal** es un conjunto de datos ordenados cronológicamente, donde cada valor depende del tiempo.  
El análisis de series temporales permite **entender la evolución de un fenómeno a lo largo del tiempo** y **predecir comportamientos futuros**.

- **Componentes principales**

1. **Tendencia (Trend):**  
   - Indica la dirección general de la serie a largo plazo.  
   - **Ejemplo:** crecimiento sostenido de las ventas a lo largo de varios años.

2. **Estacionalidad (Seasonality):**  
   - Patrón que se repite a intervalos regulares.  
   - **Ejemplo:** aumento de ventas en diciembre por la temporada navideña.

3. **Ciclo (Cycle):**  
   - Fluctuaciones de largo plazo asociadas a factores económicos o sociales.  
   - **Ejemplo:** ciclos económicos de recesión y expansión.

4. **Ruido (Noise):**  
   - Variaciones aleatorias sin patrón definido.  
   - **Ejemplo:** picos de ventas por un evento inesperado.

**Tipos de análisis**
  - **Descomposición de series temporales:** separar tendencia, estacionalidad y ruido.  
  - **Modelado predictivo:** usar modelos estadísticos o de aprendizaje automático para estimar valores futuros.  
  - **Análisis de correlaciones temporales:** detectar relaciones entre variables en distintos momentos.

**Modelos comunes**

| Tipo de modelo | Descripción breve |
|----------------|-------------------|
| **AR (Autoregressive)** | Predice el valor actual a partir de valores pasados. |
| **MA (Moving Average)** | Usa el promedio de errores pasados para suavizar la serie. |
| **ARMA (AR + MA)** | Combina autoregresión y promedio móvil. |
| **ARIMA** | Incluye diferenciación para series no estacionarias. |
| **SARIMA** | Extiende ARIMA incorporando estacionalidad. |
| **Prophet (Meta/Facebook)** | Modelo moderno flexible para tendencia y estacionalidad. |
| **LSTM (Deep Learning)** | Redes neuronales recurrentes que capturan dependencias temporales complejas. |

#### **2. Medidas de tendencia central**
---

**Concepto general**

- Las medidas de tendencia central buscan identificar el **centro** o un **valor representativo** de una distribución de datos.
- Permiten resumir la información de una muestra o población en un solo número.


**Media aritmética**

- Es la medida de tendencia central más conocida.
- Se calcula como: **suma de todos los valores / número de valores**.
- Cuando hay **valores repetidos**, se adapta usando las **frecuencias**.
- Limitación: es muy sensible a **valores atípicos (outliers)**.

$% Media aritmética
\bar{x} = \frac{\sum_{i=1}^{n} x_i}{n}$

$% Media con frecuencias
\bar{x} = \frac{\sum_{i=1}^{k} f_i x_i}{\sum_{i=1}^{k} f_i}$

**Media ponderada**

- Variante de la media en la que cada valor tiene un **peso**.
- Fórmula: cada valor se multiplica por su peso y se divide entre la suma de los pesos.
- Útil cuando algunos valores son más importantes que otros.

$% Media ponderada
\bar{x}_w = \frac{\sum_{i=1}^{n} w_i x_i}{\sum_{i=1}^{n} w_i}$

**Media armónica**

- Se emplea en situaciones como el cálculo de **velocidades medias**.
- Ningún valor puede ser cero (por la división).
- Menos utilizada que la media aritmética o ponderada.

$% Media armónica
\bar{x}_h = \frac{n}{\sum_{i=1}^{n} \frac{1}{x_i}} \quad (x_i \neq 0)$

**Otras medias**

- Existen también la **media geométrica** y la *cuadrática* (ver capítulo recomendado).

**Mediana**

- Valor que ocupa la **posición central** del conjunto de datos ordenados.
- **Si n es impar:** la mediana es el valor central.
- **Si n es par:** es el promedio de los dos valores centrales.
- Es **más robusta que la media**, pues no se ve tan afectada por los outliers.
- Útil para grandes conjuntos de datos o cuando hay valores extremos.
- Puede calcularse también en **datos agrupados**, usando una fórmula de interpolación.
- Identifica el punto donde el **área acumulada del histograma es 50%**.
- Puede emplearse con **variables cualitativas ordinales**, a diferencia de la media.

$% Mediana (datos agrupados)
Me = L_i + \frac{\frac{N}{2} - F_{i-1}}{f_i} \cdot c_i$

**Moda**

- Es el **valor más frecuente** en la distribución.
- Puede haber:
  - 1 moda (unimodal)
  - 2 modas (bimodal)
  - 3 o más modas (multimodal)
- Es la medida de tendencia central más simple y puede usarse con **variables cualitativas**.

$% Moda (datos agrupados)
Mo = L_i + \frac{f_i - f_{i-1}}{(f_i - f_{i-1}) + (f_i - f_{i+1})} \cdot c_i$

#### **3. Medidas de tendencia central robustas**
---

**Concepto general**

- La **media aritmética** es sensible a valores extremos (**outliers**).
- En presencia de outliers, la media **no es un estimador fiable** de la tendencia central.
- La **mediana** es una alternativa robusta porque **no se ve afectada** por valores extremos.
- Sin embargo, su interpretación es distinta a la de la media.

**Medidas de tendencia central robustas basadas en la media**

1. **Media recortada (trimmed mean)**
- Calcula la media aritmética **solo del subconjunto central de los datos**.
- Los valores más pequeños y más grandes se **eliminan**, reduciendo el impacto de los outliers.
- Se expresa como **“media recortada al y%”**, donde y es el porcentaje eliminado en cada extremo.
  - **Ejemplo:** Con 10 datos y un recorte del 40%, se eliminan 4 valores por la izquierda y 4 por la derecha.
- Una media recortada al **0%** es equivalente a la media normal.
- La media recortada al **25%** se denomina centrimedia.
- También se puede hablar de **niveles**: nivel 1 = recortar 1 dato por cada lado.

2. **Media winsorizada (winsorized mean)**
- Similar a la media recortada, pero **no elimina** los valores extremos.
- En lugar de eliminarlos, los **sustituye** por los valores más cercanos dentro del rango central.
  - **Ejemplo de nivel 2:** se reemplazan los 2 valores más bajos y los 2 más altos por los valores extremos del conjunto restante.
- Reduce la influencia de los outliers, pero conserva el tamaño total de la muestra.

**Objetivo de estas medidas**
- Reducir el impacto de los valores extremos.
- Calcular un valor de tendencia central que refleje mejor el **comportamiento típico** de los datos.
- Centrarse en los valores del **núcleo de la distribución**.

#### **4. Medidas de dispersión**
---

**Concepto general**
- Las **medidas de dispersión** indican cuánto se **desvían** o **separan** los datos entre sí.
- Son fundamentales para entender la **variabilidad** del conjunto de datos y complementar la información proporcionada por las medidas de tendencia central.

1. **Rango**
- Es la medida de dispersión más simple.
- **Rango = valor máximo – valor mínimo**.
- Útil, pero limitado porque solo considera dos valores de toda la distribución.

$% Rango
R = X_{\text{max}} - X_{\text{min}}$

2. **Varianza**
- Mide cuánto se **desvían los datos respecto a la media**.
- Es el **promedio de las desviaciones cuadráticas**:
  - Se elevan al cuadrado para evitar que valores positivos y negativos se cancelen.
- Varianza poblacional se denota como **$σ$²**.
- Tiene una versión para datos agrupados y muestrales.
- En tradiciones anglosajonas se divide entre **n – 1** (en lugar de n), lo cual se explica por criterios estadísticos de estimación (se verá más adelante).

$% Varianza poblacional
\sigma^2 = \frac{\sum_{i=1}^{n} (x_i - \mu)^2}{n}$

3. **Desviación típica o estándar**
- Es la **raíz cuadrada de la varianza**.
- Se usa porque devuelve la medida a la **misma unidad** que los datos.
- Siempre es **positiva**, pues mide una distancia.
- Muy utilizada para evaluar la **dispersión real** en términos comprensibles.

$% Desviación típica
\sigma = \sqrt{\sigma^2}, \quad s = \sqrt{s^2}$

4. **Coeficiente de variación (CV)**
- Permite comparar la variabilidad entre poblaciones con unidades distintas.
- Se calcula como:
  - $CV = desviaciónTípica / media$
- Es una medida **sin unidades**.
- Útil cuando:
  - Las variables están en **diferentes escalas** *(ej.: cm vs kg)*.
  - Se quiere comparar quién tiene más variabilidad relativa.

$% Coeficiente de variación
CV = \frac{s}{\bar{x}}$

**Ejemplos clave**
- **Ejemplo de Gasol:** se utiliza la fórmula adaptada para calcular la desviación típica a partir de una tabla de frecuencias.
  - **Ejemplo de peso y estatura:**
    - No pueden compararse directamente por estar en diferentes unidades.
    - El coeficiente de variación permite ver que la estatura tiene mucha menos variabilidad que el peso.

#### **5. Medidas de dispersión robustas**
---

**Concepto general**
- Al igual que la media, la **varianza** y la **desviación típica** son sensibles a los **outliers**.
- Para solucionar esto, existen versiones **robustas** basadas en el proceso de **winsorización**.

1. **Varianza winsorizada**
- Se calcula aplicando la **winsorización** a los datos antes del cálculo.
- Los datos extremos se **sustituyen** por los valores más cercanos dentro del rango central (siguiendo el mismo proceso usado para la media winsorizada).
- **Fórmula:** se utiliza el conjunto de datos winsorizado (**Wi**) y la **media winsorizada** correspondiente.
- Reduce el impacto de valores atípicos en la estimación de la dispersión.

$% Varianza winsorizada
\sigma_W^2 = \frac{\sum_{i=1}^{n} (W_i - \bar{W})^2}{n}$

2. **Cuasivarianza winsorizada**
- Es la versión winsorizada de la **cuasivarianza** (o varianza muestral).
- Sigue la misma fórmula que la varianza winsorizada, pero divide entre **n − 1** en lugar de n.
- Se usa cuando trabajamos con muestras y se desea un estimador insesgado.

$% Cuasivarianza winsorizada
s_W^2 = \frac{\sum_{i=1}^{n} (W_i - \bar{W})^2}{n-1}$

3. **Cuasidesviación típica winsorizada**
- Es la **raíz cuadrada** de la cuasivarianza winsorizada.
- Proporciona una medida robusta de dispersión en las **mismas unidades** que los datos.
- Funciona igual que la desviación típica convencional, pero sin verse afectada por valores extremos.

$% Cuasidesviación típica winsorizada
s_W = \sqrt{s_W^2}$

#### **6. Medidas de posición y forma**
---

1. **Medidas de posición**

**Tipificación (z-scores)**
- Permite saber cuán lejos está un valor respecto a la media.
- Fórmula:
  - $𝑧𝑖=𝑥𝑖−𝑥ˉ𝑠zi	=sxi	−xˉ$​
- Interpretación:
  - Indica cuántas desviaciones típicas se aleja un valor de la media.
  - Elimina unidades → permite comparar variables con escalas diferentes.
- Regla general:
  - Valores entre −2 y 2 → frecuentes.
  - Valores fuera de ese rango → infrecuentes (según la distribución normal).

$% Tipificación (z-score)
z_i = \frac{x_i - \bar{x}}{s}$

**Cuartiles**
- Dividen los datos en **4 partes iguales**.
- Q1: deja debajo el 25% de los datos.
- Q2: deja el 50% (equivale a la mediana).
- Q3: deja el 75%.
- Útiles para describir la posición de los datos y detectar valores atípicos.

$% Posición de cuartiles (datos ordenados)
Q_1 = \frac{N+1}{4}, \quad Q_2 = \frac{N+1}{2}, \quad Q_3 = \frac{3(N+1)}{4}$

**Percentiles y cuantiles**
- Los cuantiles son la categoría general.
- Percentiles: dividen los datos en 100 partes.
  - Ej: P40 → deja debajo el 40% de los datos.
- **Deciles:** dividen en 10 partes (9 deciles).
- Relación:
  - Q1 = P25
  - Q2 = P50 = D5 = Mediana
  - Q3 = P75

**Cálculo de cuartiles con datos ordenados**
- La posición se calcula con:
  - P25 = (N+1)/4
  - P50 = (N+1)/2
  - P75 = 3(N+1)/4
- Si la posición no es un entero → se toma el promedio entre los dos valores vecinos.

**Cuartiles en datos agrupados**
- Igual proceso que la fórmula de la mediana para datos agrupados.
- Se utiliza el intervalo cuantil (el intervalo que contiene el percentil deseado).
- Fórmula general (análogo a la mediana agrupada).

$% Interpolación de cuartiles para datos agrupados
Q_i = L_i + \frac{P_i - F_{i-1}}{f_i} \cdot c_i$

2. **Medidas de forma**

**Simetría**
- La distribución puede ser:
  - **Simétrica**: media = mediana = moda.
  - **Asimetría positiva (cola a la derecha):** media > mediana > moda.
  - **Asimetría negativa (cola a la izquierda):** moda > mediana > media.
- Se puede medir mediante el **coeficiente de asimetría de Pearson**:
  - Positivo → asimetría positiva.
  - Cero → simetría.
  - Negativo → asimetría negativa.

$% Coeficiente de asimetría de Pearson
A_s = \frac{\bar{x} - Mo}{s}$

**Curtosis (apuntamiento)**
- Indica si la distribución es más o menos "picuda" que la normal:
  - **Platicúrtica:** más achatada.
  - **Mesocúrtica:** apuntamiento normal (como la normal).
  - **Leptocúrtica:** más puntiaguda.

#### **7. Gráficos de caja (Boxplot)**
---

1. **Concepto**

- Permiten resumir la distribución de datos en un gráfico sencillo.
- Utilizan cinco medidas clave:
  1. Mínimo
  2. Primer cuartil (Q1)
  3. Mediana (M)
  4. Tercer cuartil (Q3)
  5. Máximo
- A partir de Q1 y Q3 se calcula el rango intercuartílico (RIQ):
  - $RIQ=Q3−Q1$
- $RIQ$ indica el intervalo donde se encuentra el 50% central de los datos.

2. **Construcción del diagrama de caja**
  1. La caja se extiende entre Q1 y Q3.
  2. La mediana divide la caja en dos partes.
  3. Los bigotes: se extienden desde el mínimo hasta el máximo, excepto valores atípicos.
    - Valores extremos: >1,5×RIQ por encima de Q3 o debajo de Q1.
    - Estos valores se marcan con círculos o asteriscos, según el software.

3. **Interpretación**
- La mediana muestra el centro de los datos.
- La longitud de la caja refleja la dispersión intercuartílica.
- Bigotes y valores atípicos permiten identificar extremos y outliers.
- Permite comparar simetría y concentración de los datos visualmente.
```r
# Lista de notas de 10 alumnos
NotasAlumnos <- c(5, 7, 8, 6, 9, 10, 6, 8, 7, 9)

# Crear diagrama de caja
boxplot(NotasAlumnos,
        main="Diagrama de caja de Notas de Alumnos",
        ylab="Puntaje",
        col="lightblue")
```

#### **8. Datos atípicos y análisis exploratorio de datos**
---

1. **Definición de valores atípicos**
- Los **valores atípicos** o **extremos (outliers)** son datos que se **alejan significativamente de la mayoría de los datos**.
- En el **diagrama de caja y bigotes**:
  - **Puntos (•) →** atípicos simples
  - **Estrellas (*) →** valores extremos

2. **Clasificación según el rango intercuartílico (RIQ)**
- Atípicos simples: entre 1,5 × RIQ y 3 × RIQ del Q1 o Q3.
- Valores extremos: > 3 × RIQ del Q1 o Q3.
- **Nota:** muchos softwares, como SPSS, hacen esta distinción, pero todos pueden considerarse simplemente outliers.
  - $RIQ=Q3−Q1$

3. **Efecto de los outliers**
- Afectan **drásticamente a la media**.
- La **mediana** apenas se ve afectada.
- Aumentan la **desviación típica** y, por tanto, la dispersión.
- Pueden **alterar la forma** de la distribución de datos (asimetría y curtosis).

4. **Análisis exploratorio de datos (EDA)**
- Es el estudio de las **características principales de un conjunto de datos** antes de hacer análisis más complejos.
- Se basa en:
  - **Medidas estadísticas:** tendencia central, dispersión, posición, forma.
  - **Gráficas:** histogramas, diagramas de caja, gráficos de dispersión, etc.
- **Objetivo:** **identificar patrones, outliers y características generales** del conjunto de datos.
