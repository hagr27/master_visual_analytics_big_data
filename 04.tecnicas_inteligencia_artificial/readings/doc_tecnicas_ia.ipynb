*El conocimiento almacenado puede:*
    - Ser **aprendido** mediante técnicas de aprendizaje automático.
    - Ser ***aportado* por un experto humano**.

*Otros puntos importantes:*

    - aprendizaje automático

- Artículo “Computing machinery and intelligence” (1950).

#### **4. Definición de aprendizaje y tipos de contenidos**

**4.1. Definición de aprendizaje (Mitchell)**

Un programa de ordenador aprende de la experiencia $E$ respecto a una clase de tareas $T$ y una medida de rendimiento $P$, si su rendimiento en $T$, medido por $P$, **mejora con la experiencia** $E$.

*Ejemplos:*
- **Ejemplo 1 – Robos de tarjetas de crédito**
    - $T$: detectar robos de tarjetas de crédito.
    - $P$: porcentaje de robos detectados.
    - $E$: base de datos de hábitos de compra con la tarjeta.
- **Ejemplo 2 – Reconocimiento de escritura manual**
    - $T$: reconocer y clasificar palabras en imágenes.
    - $P$: porcentaje de palabras correctamente clasificadas.
    - $E$: base de datos de imágenes de palabras manuscritas clasificadas.
- **Ejemplo 3 – Aparcar un coche con visión**
    - $T$: aparcar un coche usando sensores de visión.
    - $P$: porcentaje de aparcamientos correctos.
    - $E$: secuencias de imágenes y comandos de guiado registrados.

**4.2. Qué se puede aprender**

Según la “Component Display Theory”:

- **Hechos:** afirmaciones simples (ej. fecha–evento, nombre–objeto).
- **Conceptos:** conjuntos de objetos/símbolos/eventos que comparten características y se referencian por un nombre o símbolo.
- **Procedimientos:** secuencias de acciones para lograr un objetivo.
- **Principios:** relaciones causa–efecto, leyes generales.

Los ordenadores son especialmente eficaces aprendiendo conceptos representados como:
- **Árboles de decisión.**
- **Reglas.**
- **Funciones matemáticas.**

#### **5. Elementos del aprendizaje de conceptos**

*Definiciones:*
- **Instancia:** ilustración específica de un objeto, símbolo, evento, proceso o procedimiento.
 - **Ejemplo:** instancia que es miembro del concepto/clase.
- **No-ejemplo:** instancia que no es miembro del concepto/clase.
- **Atributos:** características de las instancias.
- **Clase / Concepto:** conjunto de instancias que comparten atributos comunes.
- **Datos de entrenamiento:** conjunto de instancias (ejemplos y no-ejemplos) que constituyen la experiencia $E$ para aprender la tarea.

*Ejemplo: detección de transacciones fraudulentas con tarjeta:*
- Instancia: cada transacción.
- Ejemplo: transacción fraudulenta (si se aprende “fraude”).
- No-ejemplo: transacción legal.
- Datos de entrenamiento: porcentaje de transacciones (fraudulentas y legales) usadas para aprender.

#### **6. Clasificación, generalización y discriminación**

*Una vez aprendido un concepto, se puede:*

- **Clasificación:** identificar la clase de un objeto/símbolo/evento concreto (decir de qué clase es o no es).
- **Generalización:** asignar una instancia **desconocida** a una clase observando **atributos comunes** con ejemplos previos de esa clase.
- **Discriminación:** identificar que una instancia no pertenece a una clase, al no compartir atributos con ejemplos de esa clase.

*Ejemplo ilustrativo:*

- **Clases:** **“doméstico”** vs **“salvaje”**.
- **Generalización:** clasificar un nuevo animal como **“salvaje”** si comparte atributos con animales salvajes previos.
- **Discriminación:** decidir que un animal creado en granja no es **“salvaje”**.

#### **7. Aprendizaje como búsqueda, bias y sobreajuste**

El aprendizaje de conceptos se plantea como una **búsqueda en un espacio de hipótesis** (posibles descripciones del concepto) para encontrar la que mejor se ajusta a los datos de entrenamiento.

**Hipótesis del aprendizaje inductivo de conceptos:**<br>
Cualquier hipótesis que encaje “suficientemente bien” con un conjunto “suficientemente grande” de ejemplos de entrenamiento, también encajará bien con **instancias nuevas**.

**Retos:**
- Definir un espacio de hipótesis que **incluya la mejor solución** pero no sea inabarcable.
- Determinar el tamaño adecuado del conjunto de datos de entrenamiento.

**Bias (sesgo inductivo):**
- Son los **criterios y supuestos** para seleccionar hipótesis y generalizar.
- En la práctica puede haber varias descripciones “razonablemente buenas” del concepto.

**Sobreajuste (overfitting):**
- Ocurre cuando la hipótesis encontrada es **demasiado específica** y se ajusta demasiado a los datos de entrenamiento, **generalizando mal** a nuevas instancias.
- Para detectarlo, se reserva un **conjunto de instancias** que **no participa en el entrenamiento** y se usa para **validar** la generalización.

#### **8. Aprendizaje supervisado**

*Objetivo:*
- **Caracterizar o describir un concepto/clase** a partir de instancias etiquetadas (ejemplos y no-ejemplos).

*Etapas (Figura 4):*
1. **Descripción del concepto:** construir un **modelo descriptivo** a partir de instancias etiquetadas.
2. **Clasificación de nuevas instancias:** usar el modelo para asignar una clase a instancias nuevas.

**Ejemplo médico:**

- Datos (Tabla 1): pacientes con síntomas (fiebre, dolor de garganta, congestión, dolor de cabeza) y diagnóstico.
- Atributos de entrada: síntomas.
- Atributo de salida / clase: diagnóstico.
- Con un algoritmo de árboles de decisión (ej. ID3) se aprende un modelo que:
    - Es fácilmente interpretable.
    - Se puede traducir a reglas SI–ENTONCES.

Los datos usados para construir el modelo son los **datos de entrenamiento**.
Se pueden usar **datos de prueba** adicionales para comprobar la capacidad de generalización del modelo aprendido.

#### **9. Aprendizaje no supervisado**

*Objetivo:*

- **Caracterizar un concepto desconocido** a partir de instancias **sin clases predefinidas**.
- Es una tarea de **formación de conceptos**.

*Herramienta típica:*
- **Clustering (agrupamiento):**
    - Agrupa instancias según un criterio de **similitud**.
    - Los datos de entrenamiento **no indican** qué se quiere aprender (los grupos).

**Usos:**
- Análisis exploratorio de datos (visualización de clusters).
- Descubrir **segmentos**: por ejemplo, segmentación de clientes en investigaciones de mercado.
- Generalización: incorporar nuevas instancias al cluster más similar, heredando las conclusiones asociadas al grupo.

#### **10. Etapas del aprendizaje de un concepto**

1. **Selección del conjunto de datos de entrenamiento**
    - Analizar los datos disponibles.
    - Elegir instancias relevantes.
    - La cantidad y calidad de instancias es clave.
    - Puede haber **retroalimentación directa** (clase conocida por instancia) o indirecta (la información sobre el resultado es global, por ejemplo, ganar o perder una partida).
2. **Selección de la función objetivo y su representación**
    - Definir **qué** conocimiento se va a aprender (función objetivo).
    - Elegir una **representación**: árbol de decisión, reglas, funciones, etc.
3. **Selección del algoritmo de aprendizaje**
    - Elegir el método que aproxima la función objetivo (por ejemplo, método de construcción de árboles de decisión).
    - Depende del tipo de datos, si hay ruido, si son discretos/continuos, etc.
4. **Evaluación y validación de resultados**
    - Verificar si el rendimiento es adecuado.
    - Decidir si es necesario volver a pasos anteriores (proceso cíclico) para ajustar datos, función objetivo o algoritmo.

#### **11. Descubrimiento de Conocimiento en Bases de Datos (KDD)**

El objetivo de la minería de datos se enmarca en el proceso más amplio de **Descubrimiento de Conocimiento en Bases de Datos (KDD)**:
- KDD = procedimiento completo para extraer **conocimiento útil y previamente desconocido** a partir de los datos.
- Es un **proceso iterativo** que:
    - Empieza identificando el **objetivo de la aplicación** (problema, hipótesis, resultados esperados).
    - Incluye etapas previas y posteriores a la minería de datos.

*Etapas típicas (Figura 8):*
1. **Integración de datos**
- Combinar varias fuentes (ficheros planos, BD relacionales, data warehouse).
2. **Selección de datos**
- Elegir atributos e instancias **relevantes**.
- Se apoya en conocimiento de dominio y, si procede, en herramientas automáticas.
- **Ejemplo:** eliminar atributos irrelevantes como “Número de paciente” en la tabla de diagnóstico.
3. **Pre-procesamiento de datos**
- Gestión de datos **ruidosos, inconsistentes o faltantes**.
- Limpieza de datos, detección de duplicados, outliers, etc.
- **Riesgo:** eliminar ejemplos de clases poco frecuentes.
4. **Transformación de datos**
- Normalización, conversión entre formatos:
    - Nominal → numérico (para ciertos métodos).
    - Numérico → discretizado (para otros métodos que funcionan mejor con valores discretos).
5. **Minería de datos**
- Aplicación de **técnicas de IA / AA** para extraer patrones.
6. **Interpretación y evaluación de resultados**
- Analizar si los patrones son útiles e interesantes.
- Puede implicar repetir etapas previas con nuevos atributos o instancias.
- Se apoyan técnicas de visualización y representación de conocimiento.
7. **Aplicación del conocimiento**
- Usar el conocimiento descubierto en el problema real.
- Esta etapa también sirve para validar las conclusiones obtenidas.

#### **3. Definición de aprendizaje, tareas básicas y ejemplos**

**3.1. Definición de aprendizaje en IA**

Un programa de ordenador aprende de la siguiente manera:
> *Un programa de ordenador aprende de la experiencia E con respecto a una clase de tareas T y una medida de rendimiento P, si su rendimiento en las tareas T, medido en base a la medida P, mejora con la experiencia E.
(Mitchell, 1997)*

**Ejemplos de tareas de aprendizaje**

| Ejemplo | Tarea (T)                               | Medida de rendimiento (P)                | Experiencia (E)                       |
| ------- | --------------------------------------- | ---------------------------------------- | ------------------------------------- |
| 1       | Detectar robos de tarjetas de crédito   | % de robos detectados                    | Base de datos de hábitos de compra    |
| 2       | Reconocer escritura manual              | % de palabras correctamente clasificadas | Base de datos de imágenes manuscritas |
| 3       | Aparcar un coche con sensores de visión | % de aparcamientos correctos             | Secuencias de imágenes y comandos     |

**Contenidos que se pueden aprender**

*Según la Teoría de Presentación de Componentes (Merril et al., 1994):*

- **Hechos:** afirmaciones simples (e.g., fecha de un evento, nombre de un objeto).
- **Conceptos:** agrupación de objetos, símbolos o eventos con características comunes.
- **Procedimientos:** acciones consecutivas para alcanzar un objetivo.
- **Principios:** relaciones causa-efecto o verdades generales.

> *Los ordenadores son especialmente eficaces aprendiendo conceptos, representables mediante árboles, reglas o funciones matemáticas.*

**3.2. Elementos que intervienen en el aprendizaje de un concepto**

- **Experiencia (E):** conjunto de datos (transacciones, imágenes, secuencias).
- **Instancia:** cada ejemplo concreto dentro de la experiencia.
- **Atributos:** características que definen un concepto.
- **Clase o concepto:** conjunto de instancias con características comunes.
- **Ejemplos:** instancias que pertenecen al concepto.
- **No-ejemplos:** instancias que no pertenecen al concepto.
- **Datos de entrenamiento:** todas las instancias (ejemplos y no-ejemplos) utilizadas para aprender.



**Aprender un concepto implica:**

- **Clasificación:** identificar la clase de un objeto.
- **Generalización:** aplicar la clase a instancias desconocidas según atributos comunes.
- **Discriminación:** identificar instancias que no pertenecen a la clase.

***Ejemplo:** concepto “animal salvaje”*

- **Generalización:** clasificar un animal desconocido como salvaje si vive en la jungla.
- **Discriminación:** no clasificarlo como salvaje si está en una granja.



**3.3. Tipos de tareas de aprendizaje**

- **Aprendizaje supervisado:**
  - Se aprende a partir de ejemplos etiquetados.
  - **Ejemplo:** clasificar animales domésticos basándose en atributos conocidos.
  - **Herramientas:** árboles de decisión, reglas SI-ENTONCES.

- **Aprendizaje no-supervisado:**

  - **Técnica típica:** clustering, que agrupa instancias según similitud.
  - **Aplicación:** segmentación de clientes, agrupación de alumnos con características similares.

**3.4. El aprendizaje como búsqueda**

- El aprendizaje inductivo busca la mejor hipótesis dentro de un espacio de soluciones posibles.
- **Problemas comunes:**
  - **Bias inductivo:** supuestos que permiten generalizar.
  - **Sobreajuste (overfitting):** la hipótesis se ajusta demasiado a los datos de entrenamiento y no generaliza bien.
  - **Estrategia:** reservar datos de prueba para validar la generalización.

> *El bias se refiere a los criterios de selección de las hipótesis y aquellos supuestos y suposiciones que se realizan para generalizar la mejor hipótesis encontrada.*

#### **4. Etapas en el descubrimiento de conocimiento**

En el marco del máster actual ser capaz de aplicar técnicas de IA para descubrir conocimiento en bases de datos es de gran interés, es lo que se ha definido anteriormente como minería de datos. El término **Descubrimiento de Conocimiento en Bases de Datos**, en inglés **Knowledge Discovery in Databases (KDD)**, es frecuentemente utilizado como sinónimo de la minería de datos (Fayyad, 2001; Fayyad et al., 1996).

Sin embargo, este término se refiere al **procedimiento completo** necesario para extraer conocimiento potencialmente útil y previamente desconocido a partir de los datos en una base de datos (Roiger, 2003). KDD es un proceso iterativo que incluye etapas previas a la fase de minería de datos propiamente dicha, para la extracción y preparación de los datos, así como etapas posteriores para el análisis de los resultados y toma de decisiones.

Hay diferentes modelos KDD definidos, pero sea cual sea el modelo KDD utilizado, el **primer paso** siempre consistirá en **identificar el objetivo de la aplicación** del procedimiento, esto es, definir el problema identificando la tarea que se ha de realizar, estableciendo hipótesis sobre los resultados deseados y posibles

#### **4. Medidas de impureza y selección de atributos**

**La selección de atributos**
- Busca hacer la mejor división (Best Split).
- Esta será la que separe los datos en subconjuntos lo más homogéneos posibles.
- Se utilizan las siguientes técnicas: 

**4.1. Entropía**
- Mide el **grado de incertidumbre o impureza** de los datos.
- Se usa para medir la **homogeneidad**:
    - Entropía cercana a **0** → muestra muy **homogénea** (clases iguales).
    - Entropía cercana a **1** → muestra **heterogénea** (clases mezcladas).
- Interpretación:
    - Entropía ≈ 0 → “los datos del nodo son de la misma clase”.
    - Entropía ≈ 1 → “los datos del nodo son de distintas clases”.

**4.2. Índice de Gini**
- Mide la **probabilidad de clasificar incorrectamente** una instancia.
- También mide **homogeneidad**:
    - Gini cercano a **0** → muestra homogénea.
    - Gini cercano a **1** → muestra heterogénea.

**Fórmula (en la transparencia):**



**4.3. Ganancia de información**
- Mide la **impureza (entropía) después de una división**.
- Mide cuánto se **reduce la entropía antes y después** de dividir:
    - Ganancia de información **alta** → datos más homogéneos tras la división.
    - Ganancia de información **baja** → datos siguen heterogéneos.
- Interpreta cuánta **información aporta un atributo** sobre la variable objetivo.
- **Fórmula (simplificada en la diapositiva):**

> (conceptualmente: entropía antes – entropía ponderada después).

**4.4. ¿Qué medida elegir?**
- **Índice de Gini:**
    - Mejor para distribuciones de datos **grandes**.
    - Más **eficiente computacionalmente**.
    - Tiende a favorecer divisiones que **aislan la clase más frecuente**.
- **Ganancia de información:**
    - Tiende a crear nodos **más puros**.
    - Más **costosa** computacionalmente.
- Estas medidas suelen venir ya **incorporadas** en las implementaciones de los algoritmos.

#### **5. Sobreajuste y poda en árboles de decisión**

**5.1. Sobreajuste (overfitting)**
- Ocurre cuando el modelo se ajusta **demasiado** a los datos de entrenamiento.
- Consecuencias: mala generalización a datos nuevos.
- Técnicas para mitigarlo:
    - **Reducción de dimensionalidad**.
    - Usar **algoritmos más simples**.
    - Utilizar **validación cruzada**.



> <mark> A mayor complejidad del algoritmo, mayor probabilidad de incurrir en **overfitting**

**5.2. Prepoda**
- Se aplica **durante** la construcción del árbol.
- Consiste en establecer **criterios/umbrales** que limitan la expansión:
- **Hiperparámetros**
    - **Min_samples_leaf:** nº mínimo de observaciones en un nodo hoja (terminal).
    - **Min_samples_split:** nº mínimo de observaciones para dividir un nodo.
    - **Max_features:** nº máximo de variables a considerar en cada división.
    - **Max_depth:** profundidad máxima del árbol (nº de niveles).
- Los más usados: **Max_depth** y **Min_samples_split**.



**5.3. Post-poda**
- Se aplica **después** de crear el árbol completo.
- Consiste en **eliminar nodos** (subárboles).
- Para elegir el mejor subárbol se usa **validación cruzada**.
- Técnicas principales:
    - **Poda de costo-complejidad (CCP)**.
    - **Poda de error reducido (REP)**.
    - **Poda basada en error mínimo (MEP)**.
El algoritmo **CART** utiliza **CCP**.

- El atributo con **mayor capacidad de clasificación** se convierte en nodo raíz.
- Algoritmos relacionados:
    **ID3** (Quinlan, 1986).
    **C4.5** (Quinlan, 1993).
    **CART** (Breiman et al., 1984).
- Implementaciones en librerías:
    - **scikit-learn:** https://scikit-learn.org/stable/modules/tree.html
    - **TensorFlow Decision Forests:** https://www.tensorflow.org/decision_forests/api_docs/python/tfdf/keras/CartModel

#### **1. Recordatorio rápido: árboles de decisión**

- **Aprendizaje de árboles de decisión**
    - Aprendizaje supervisado.
    - Clasificación y regresión.
    - Especialmente adecuados para variables categóricas.
- **Estructura**
    - **Nodos:** conjunto/subconjunto de datos.
    - **Ramas:** condiciones (normalmente True/False o particiones por valor).
    - **Nodo raíz**, nodos internos y nodos hoja.
    - **Objetivo:** nodos lo más homogéneos posible (misma clase).
- **Selección de atributos (Best Split)**
    - **Entropía:** homogeneidad de los datos.
        - Entropía ≈ 0 → datos homogéneos.
        - Entropía ≈ 1 → datos heterogéneos.
        - Usada en ID3 y C4.5, y en Ganancia de información.
    - **Índice de Gini:**
        - Probabilidad de clasificación incorrecta.
        - Gini ≈ 0 → homogéneo, Gini ≈ 1 → heterogéneo.
        - Usado en CART.
    - **Ganancia de información:**
        - Reducción de incertidumbre tras la partición.
        - Usa entropía.
        - Ganancia alta → nodo más homogéneo.
        - Usada en ID3 y C4.5.
    - **Otras:** Proporción de ganancia de información (junto con Ganancia de información).
- **Sobreajuste u overfitting**
    - El árbol se ajusta demasiado a los datos de entrenamiento.
    - **Consecuencias:** mala generalización.
- **Técnicas para reducir sobreajuste (recordatorio)**
    - Validación cruzada (cross-validation).
    - Técnicas de ensemble.
    - Pre-poda y post-poda.
- **Pre-poda: limitar crecimiento con hiperparámetros**
    - **min_samples_leaf:** nº mínimo de muestras en una hoja.
    - **min_samples_split:** nº mínimo de muestras para dividir un nodo.
    - **max_features:** nº máximo de variables consideradas en cada split.
    - **max_depth:** profundidad máxima del árbol.
    - **Los hiperparámetros más usados:** max_depth y min_samples_split.

#### **2. Poda de árboles y validación cruzada**

**2.1. Poda de árboles**

- **Definición**
    - Reducir el tamaño del árbol una vez construido.
    - **Objetivo:** mejorar generalización y reducir sobreajuste.
- **Tipos**
    - **Poda directa:**
        - Elimina subárboles completos.
        - Sustituye subárbol por un nodo hoja.
        - Se puede realizar con técnicas como CCP (cost-complexity pruning).
    - **Poda por reglas:**
        - Se obtienen reglas a partir del árbol.
        - Se simplifican eliminando condiciones de las reglas.
    - **Cómo decidir el tamaño adecuado del árbol**
        - Validación cruzada (cross-validation).
        - Poda pesimista de C4.5.
        - Simplificación mediante poda de reglas.

**2.2. Validación cruzada (Cross-Validation)**

- **Objetivos**
    - Evaluar la capacidad de generalización del modelo en datos nuevos.
    - Ayudar a evitar el sobreajuste.
- **Esquema k-fold**
    - Dividir datos en k subconjuntos (folds), típicamente 5 o 10.
    - Para cada iteración:
        - 1 fold → test.
        - k-1 folds → entrenamiento.
    - Repetir k veces, cambiando el fold de test.
    - **Métrica final:** promedio sobre las k iteraciones.

#### **3. Medidas de precisión de la clasificación**

**3.1. Matriz de confusión**
  - **Estructura (binaria)**
    - Observación vs Predicción:
        - **VP (TP):** verdaderos positivos.
        - **VN (TN):** verdaderos negativos.
        - **FP:** falsos positivos.
        - **FN:** falsos negativos.
        - Las predicciones correctas están en la diagonal (VP y VN).
        - Válida para clasificación binaria y multiclase.

**3.2. Métricas derivadas**
- **Accuracy**
    - Proporción de predicciones correctas sobre el total.
    - 
- **Precision**
    - De los ejemplos predichos como positivos, cuántos son realmente positivos.
    - 
- **Recall (sensibilidad)**
    - De los positivos reales, cuántos se predicen correctamente.
    - 
- **F-measure (F1)**
    - Media armónica de precisión y recall.
    - 

**3.3. Ejemplo interpretado (riesgo cardiovascular)**
- **Matriz (resumen):**
  - **TN** = 390, FP = 13, FN = 24, TP = 23 (total 450).
- **Cálculos:**
  - **Accuracy** = 413 / 450 ≈ 0,917.
  - **Precision** = 23 / 36 ≈ 0,63.
  - **Recall** = 23 / 47 ≈ 0,48.
- **Interpretación:**
  - **Accuracy 0,91:** el modelo acierta en el 91% de los casos.
  - **Precision 0,63:** de los pacientes predichos con riesgo, el 63% realmente lo tienen (37% FP).
  - **Recall 0,48:** de los que realmente tienen riesgo, el modelo solo detecta el 48% (52% FN).

**3.4. Curva ROC-AUC**
- **ROC (Receiver Operating Characteristic)**
  - **Eje Y:** TPR (True Positive Rate = Recall).
  - **Eje X:** FPR (False Positive Rate = FP / (FP + TN)).
  - Muestra compromiso entre TPR y FPR para distintos umbrales.
- **AUC (Area Under Curve)**
  - **AUC** = 0,5 → capacidad predictiva baja (similar a clasificador aleatorio).
  - **AUC** = 1 → capacidad óptima.
- **Diferencia con matriz de confusión**
  - **Matriz de confusión:** usa un umbral concreto y da VP, VN, FP, FN.
  - **ROC-AUC:** resume comportamiento del clasificador en todos los umbrales, centrado en TP y FP.

#### **4. Ensemble Learning y Random Forest**

**4.1. Métodos de ensemble**
- **Objetivo**
  - Mejorar la capacidad predictiva combinando varios modelos.
- **Tipos principales**
  - Boosting:
    - Modelos entrenados secuencialmente.
    - Cada modelo se centra en los errores del anterior.
    - **Ejemplos:** AdaBoost, Gradient Boosting.
  - Bagging:
    - Modelos entrenados en paralelo.
    - Cada modelo se entrena con una muestra bootstrap del dataset.
    - Se combinan predicciones (votación / promedio).

**4.2. Random Forest**
- **Características**
  - Método de ensemble basado en árboles de decisión.
  - Usa bagging y aleatoriedad en la selección de variables.
  - **Cada árbol:**
    - Se entrena con un subconjunto bootstrap de los datos.
    - En cada split se consideran aleatoriamente solo algunas variables.
  - **Objetivo:** decorrelacionar los árboles y mejorar generalización.
- **Etapa 1: Creación de conjuntos de datos (bootstraping)**
  - **Para cada árbol:**
    - Se seleccionan instancias con reemplazo (mismo tamaño que la muestra original).
    - Las instancias no seleccionadas se pueden usar como out-of-bag (para evaluar).
  - Se busca variabilidad entre árboles.
- **Etapa 2: Entrenamiento de cada árbol**
  - Cada árbol se entrena con su subconjunto de datos.
  - En cada división, se selecciona aleatoriamente un subconjunto de variables.
  - **Diferencia clave:**
    - Árbol de decisión “clásico”: usa todas las variables disponibles en cada split.
    - Random Forest: introduce aleatoriedad en las variables consideradas.
- **Etapa 3: Obtención de la predicción**
  - Cada árbol predice de forma independiente.
  - Combinación:
    - Clasificación: votación por mayoría (clase más frecuente).
    - Regresión: promedio de predicciones.
- **Implementaciones mencionadas**
  - Forests of randomized trees en scikit-learn:
    - https://scikit-learn.org/stable/modules/ensemble.html#forests-of-randomized-trees
  - TensorFlow Decision Forests:
    - https://www.tensorflow.org/decision_forests?hl=es-419
    - https://keras.io/examples/structured_data/classification_with_tfdf/

#### **5. Reglas de clasificación y reglas de asociación**

**5.1. Propiedades deseables de la representación del conocimiento**
  - Comprensible.
  - Eficiente.
  - Adaptable.
  - Consistente (gestiona redundancias y conflictos).
  - Buena cobertura.
  - Completitud.
- **Ventajas de los sistemas de reglas**
  - Modularidad.
  - Fácil ampliación y modificación.
  - Fáciles de entender.
  - Separación entre control y conocimiento.
  - Permiten explicar decisiones.
- **Estructura de una regla**
  - **Antecedente:** condiciones de aplicación.
  - **Consecuente:** conclusión / acción.
  - **Sintaxis:**
    - SI <antecedente> ENTONCES <consecuente>
  - **Puede haber:**
    - Múltiples condiciones unidas por AND / OR.
    - Múltiples conclusiones en el consecuente.
    - **Recomendación:** no mezclar AND y OR en la misma regla.
- **Antecedentes y consecuentes**
  - **Antecedente:** pares objeto–valor, con operador:
    - Numérico (>, <, =, …) o lingüístico.
  - **Consecuente:**
    - Valores lingüísticos.
    - Valores numéricos o expresiones aritméticas (ej.: riesgo = edad * 1.5).
- **Relación árboles ↔ reglas**
  - Árboles de decisión ↔ conjunto de reglas de clasificación (mapeo posible en ambas direcciones).
  - Árbol → reglas:
    - Cada camino desde raíz hasta una hoja → una regla.
    - **Antecedente:** restricciones de ramas.
    - **Consecuente:** clase de la hoja.
  - Útil cuando el árbol es grande y difícil de interpretar.
- **Diferencia clave**
  - **Reglas de clasificación:**
    - Predicen la clase.
  - **Reglas de asociación:**
    - Predicen valores de atributos, combinaciones atributo-valor, o la propia clase.

#### **6. Medidas para evaluar reglas: confianza, soporte, cobertura, lift**

- **Confianza (Confidence)**
  - Probabilidad condicional de B dado A.
  - 
- Interpretación: porcentaje de ejemplos que cumplen la regla (antecedente y consecuente) respecto a los que cumplen el antecedente.
- **Soporte (Support)**
  - Proporción de ejemplos que cumplen antecedente y consecuente sobre el total.
  - 
- Sirve para descartar reglas que, aunque tengan alta confianza, se basan en muy pocos casos.
- **Cobertura**
  - Nº de ejemplos que cumplen la regla (antecedente y consecuente).
  - Se usa tanto para reglas como para item-sets.
- Ejemplo **“Jugar al aire libre” (resumen)**
  - Datos de ambiente, temperatura, humedad, viento, y clase (jugar sí/no).
  - Ejemplos de reglas de asociación:
    - Regla 1: SI Ambiente = nublado ENTONCES jugar = sí
    - Regla 2: SI Temperatura = baja ENTONCES Humedad = normal
    - Regla 3: SI Temperatura = media ENTONCES Humedad = alta
  - Cálculo:
    - Confianza regla 1 = 4/4 = 1, soporte = 4/14 ≈ 0,29.
    - Confianza regla 2 = 4/4 = 1, soporte = 4/14 ≈ 0,29.
    - Confianza regla 3 = 4/6 ≈ 0,67, soporte = 4/14 ≈ 0,29.
- **Lift**
  - Relación entre probabilidad de que se cumpla B dado A y probabilidad de B.
  - 
- **Interpretación:**
  - lift = 1 → A y B independientes.
  - lift > 1 → correlación positiva; A implica probablemente B; regla útil.
  - lift < 1 → correlación negativa; A implica probablemente no B; regla no útil.
- **Proceso típico de selección de reglas de asociación**
  1. Encontrar reglas con soporte ≥ umbral mínimo.
  2. De esas, seleccionar las que tienen confianza ≥ umbral mínimo.
  - Es necesario controlar ambos (confianza y soporte).

#### **7. Algoritmos de aprendizaje de reglas de clasificación**

**7.1. Recubrimiento secuencial (sequential covering)**
- **Idea general**
  - Se construyen reglas de forma iterativa para cada clase.
- **Esquema básico (por clase C)**
  1. Inicializar conjunto de ejemplos E.
  2. Mientras queden ejemplos de clase C en E:
    - Aprender una regla que cubra algunos ejemplos de clase C.
    - Añadir la regla al conjunto de reglas.
    - Eliminar de E los ejemplos correctamente clasificados por esa regla.
Cobertura
nº de ejemplos que cumple la regla (antecedente + consecuente).
Se puede usar para decidir reglas a mantener o descartar.
7.2. Algoritmo de aprendizaje de una regla (búsqueda greedy)
Características
Búsqueda codiciosa (greedy), sin retroceso.
De lo general a lo específico:
Empezar con regla sin restricciones en el antecedente.
Ir añadiendo parejas atributo=valor.
Mientras:
La regla cubra ejemplos negativos.
Queden atributos disponibles.
En cada iteración:
Probar añadir posibles restricciones (atributo=valor).
Medir calidad (por ejemplo, confianza).
Elegir la “mejor restricción”.
Añadirla al antecedente y eliminar ese atributo del conjunto de atributos disponibles.
7.3. Algoritmo PRISM
Tipo
Algoritmo simple de recubrimiento secuencial.
Base
Usa el algoritmo básico de recubrimiento secuencial + algoritmo de aprendizaje de una regla.
Criterio de selección de restricción
Usa la confianza como medida principal.
Confianza = (ejemplos que cumplen antecedente y consecuente) / (ejemplos que cumplen antecedente).
Estrategia (ejemplo “jugar=no”)
Paso 1:
Partir de regla general: SI --- ENTONCES jugar = no.
Evaluar confianza de restricciones como Ambiente=soleado, Humedad=alta, etc.
Elegir la de mayor confianza (ej.: Ambiente=soleado con confianza 0,60).
Paso 2:
Añadir restricción (Ambiente=soleado) y eliminar atributo Ambiente.
Volver a generar posibles restricciones con atributos restantes.
Si hay empate en confianza, elegir la de mayor cobertura.
Parar cuando la regla ya no cubre ejemplos negativos.
Regla devuelta de ejemplo:
SI Ambiente=soleado AND Humedad=alta ENTONCES jugar=no.
Otros algoritmos de reglas de clasificación mencionados
ZeroRule:
Predice siempre la clase más frecuente (modo) o la media (variable numérica).
OneRule:
Utiliza un solo atributo.
Para cada valor del atributo, asigna la clase (o valor) más frecuente.
Elige el atributo con menor error total.
RIPPER:
Basado en IREP y en REP (Reduced Error Pruning) y en Separate-and-Conquer.
Algoritmo de reglas con poda para reducir error.
