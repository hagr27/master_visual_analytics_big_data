Sí, existen sistemas multi-agente donde varios agentes se coordinan, y la salida de uno sirve como prompt (o se integra dentro de un prompt más grande) para otro agente. Como comenta tu compañero, lo habitual es que se instruya a los agentes para que las salidas sean muy estructuradas, como JSON y así puedan ser fácilmente entendidas por el otro agente, así que curiosamente acaba siendo una mezcla de información no estructurada con información semi-estructurada (los JSON).

Por cierto, te puede interesar esto:
